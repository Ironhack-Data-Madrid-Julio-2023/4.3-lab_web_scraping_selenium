```python
def extract_table_data(driver, tabela):
    table_element = driver.find_element(By.CLASS_NAME, tabela)
    table_html = table_element.get_attribute('outerHTML')
    df = pd.read_html(table_html)[0]  # Extracting table data as DataFrame
    time.sleep(3)
    
    return df

# Define your XPath and other necessary variables
tabela = "sales-insights-performance-data-table"  # Replace "tabela" with your actual XPath
num_iterations = 100  # Number of times to loop

# Create a list to store each DataFrame
data_frames = []

# Loop the function multiple times
for _ in range(num_iterations):
    data_df = extract_table_data(driver, tabela)
    # Perform any further operations with the extracted data_df
    print(data_df.head())
    
    # Add the current DataFrame to the list
    data_frames.append(data_df.copy())
    
    # Click the button to load the next page
    td_element = driver.find_element(By.CLASS_NAME, "data-table-controls__previous")
    button_element = td_element.find_element(By.CLASS_NAME, "bui-button--secondary")
    button_element.click()
    
    # Add any necessary delays between iterations
    time.sleep(3)

# Access each saved DataFrame in the list
for i, df in enumerate(data_frames):
    print(f"DataFrame {i+1}:")
    print(df.head())
    print("-------------------------------")
```
